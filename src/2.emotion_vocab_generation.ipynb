```python
from datasets import load_dataset
dataset = load_dataset('emotion')
train_data = dataset['train'] # VOCAB  is generated by training set
print(train_data[0])
```

```python
import re

def split_tokens(sentence):                             
    # split sentence into tokens/words
    return [w for w in re.split(r" +",  re.sub(r"[^a-z@# ]", "", sentence.lower()))]   

all_words = []
for sample in train_data:
    all_words = all_words + split_tokens(sample['text']) 
print('collect all words from sentences in train data, and show 10 first words: ', all_words[:10])

```

```python
from torchtext.vocab import vocab
from collections import Counter, OrderedDict

# build vocab - using vocab object of torchtext 
counter = Counter(all_words)
sorted_by_freq_tuples = sorted(counter.items(), key=lambda x: x[1], reverse=True)
my_vocab = vocab(OrderedDict(sorted_by_freq_tuples), specials=['<pad>','<unk>'])
my_vocab.set_default_index(my_vocab['<unk>'])
```

```python
my_vocab.get_itos()[:10] # this is a map - convert from id (number) to token/word 
```

```python
map = my_vocab.get_stoi()  # this is a map - convert from word to id
print(map['feel'])
print(map['feeling'])
```

```python
# method to convert list of token to the list of id 
print(my_vocab.lookup_indices(['i', 'feel', 'and', 'to', 'the', 'a', 'feeling', 'that', 'of', 'my']))
```

```python
# method to convert list of id  to the list of token 
print(my_vocab.lookup_tokens([10, 3,4,5, 9]))
```

```python
print(my_vocab(['hi','this_is_a_wrong_token']))
```
